## NetworKit Image Segmentation

This notebook can be used for image segmentation based on graph partitioning. It uses NetworKit to generate a graph from an image file and write the image segements (the partitioned graph) back to an image file.
The user can implement its on clustering algorithm or use one of the existing algorithms of NetworKit to partition the graph.

### Preparation

If you did not install NetworKit system wide, you first have to change to the directory where you installed NetworKit.

```python
from NetworKit import *
from segmentation import *
from IPython.display import Image, display
```

First, we initialize the reader and writer to convert an image to a graph and vice versa.

```python
reader = ImageGraphReader()
writer = ImageGraphWriter()
```

### Segmentation

Use the reader to convert an image into a graph. The *read* method of ImageGraphReader takes the path to the image and three additional arguments which are optional:
* *sigmaI* - Regulates the sensitivity to differences in brightness or color. Use higher values for less sensitivity.
* *sigmaX* - Regulates the impact of more distant pixels. Use higher values for less impact.
* *r* - Defines the radius of adjacent pixels to which edges are created. Edges are added to pixels with distance [0, r) from the center pixel.

If you do not provide these arguments the reader uses *sigmaI*=15, *sigmaX*=75 and *r*=4.

```python
G = reader.read("../../input/segmentation/sample_1.png", 15, 75, 4)
```

After you have converted your image into a graph, you can use it like any other NetworKit graph. For segmenting the image you might want to implement a graph clustering algorithm or use an already existing one, e.g. PLM provided by NetworKit:

```python
partition = community.detectCommunities(G)
```

To convert your graph back into an image use one of ImageGraphWriter's methods. The class provides you two methods from which you can choose:

* *writePartitionsToSingleImage*: Writes the partitions into one image, each partition colored with the mean color of its members.
* *writePartitions*: Copies the pixels in each partition from the source image and writes them into a separate image file.
  
The arguments of both methods are as follows:
* *graph* - The graph
* *partition* - The partition of your graph
* *pathToSourceImage* - The path to the source image

```python
writer.writePartitionsToSingleImage(G, partition, "../../input/segmentation/sample_1.png")
```

```python
initialImage = Image(filename="../../input/segmentation/sample_1.png")
partitionedImage = Image(filename="../../input/segmentation/sample_1_partitioned.png")
display(initialImage, partitionedImage)
```
